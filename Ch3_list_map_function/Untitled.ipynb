### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list4 = ['A','B','C','D']
print(list4)
del list4[2]
print(list4)
#print(list4[1:3])
del list4[1:3]
print(list4)
```

## Sort
+ list 進行排序有 sorted( ) 與 sort( ) 兩個動作。
+ 兩者差異：
  + sorted()
    + 不會影響 list 本身結構，且可以輸出結果。
    + sorted(list項目)。
    + sorted( ) 可用來排序任何項目。
+ sort()
    + 會影響 list 本身結構，且不可以於操作這個動作時進行輸出。
    + list.sort( )。
    + sort( ) 只能用在list上排序。

### 排序原則與參數
+ 排序原則：
  + 小到大排序
  + 數字則依照大小排序。
  + 字串則依照 ASCII 編碼順序進行排序。
  + 小寫>大寫>數字
+ 可加入的參數：
   + reverse=True  
     + 有這個參數由大到小排序
     + 若沒有這個參數則由小到大排序
+ key=len
     + 依照 list 內元素字串長度進行排序
+ key=str.upper
     + 將 list 內元素轉換為大寫，再依照 ASCII 編碼順序進行排序
+ key=str.lower
     + 將 list 內元素轉換為小寫，再依照 ASCII 編碼順序進行排序

```python
list5 = ['c','B','a']
print(list5)
list5.sort()
print(list5)

print(ord('c'))##轉換為ASCII 編號
print(ord('B'))
print(ord('a'))
list5.sort(reverse=True)
print(list5)
```

## Dictionary 介紹
+ 資料結構是由 key:value 所組成。
+ key 不能夠重複，否則會被後面的結果蓋過去。
+ 可輸入 key 找尋您要找出來的值。
+ 如果輸入的 key 不存在，那就會出現錯誤訊息。

```python
dict1 = {"a":100,100:"ken",'c':300}
print(dict1['a'])
print(dict1[100])
print(dict1["c"])
print(dict1["d"])#不存在 錯誤
```

```python
diect2 = {"a":100,'b':400,"b":300}#重複的 key覆蓋之前的value
print(diect2)
```

## key 與 value
+ key 不能於程式內改變：
  + 可以用數字、字串或者 tuple。
  + 不可以使用 list。
+ 如何找出所有的 key 與 value？
  + 您可以透過 dict1.keys( ) 這個方法找出所有的 key。
  + 您可以透過 dict1.values( ) 這個方法找出所有的 value。

## 新增與修改
+ 新增一筆資料
  + 請將 key 與 value 儲存至一個變數內。
  + 透過 dict.update(新增的資料) 這個方法的方式就可以新增。
+ 元素可否變更：
  + 以 = 指派方式指派給 key 就可以變更資料

## Function
+ 自訂函數的目的：
  + 減少相同的語法重複出現。
  + 避免重複出現的目的，是未來維護程式只要維護一個地方，不用改好幾個地方。改好幾個地方代表出錯的機率愈大。
+ 被呼叫的函數必須放在程式區塊的上方，代表已經準備好了，下方的程式區塊才可以呼叫使用。
+ 自訂函數可以放在其他檔案內，之後以模組引用 import 的方式使用：
  + 那自訂函數的內容就交給其他人負責，而你就只負責主程式的流程。
+ 在python函數是一種物件

```python
def func1():
    print("Func1")
func1()    
```

```python
def func2(x):
    ans = x**2 + 1
    print(ans)
x = func2(3)    
print(x)
```

# 單筆回傳

```python
def func3(v1,v2):
    result = v1 * v2 
    return result #離開函數 與 回傳數質
ans = func3(20,3)
print(ans)
```

# 多筆回傳

```python
def func4(v1,v2,v3):
    return (v1+2,v2+3,v3+6)
x,y,z = func4(2,3,5)
print(x,y,z)
ans = func4(2,3,5)
print(ans)
```
