## 輸入四個月的支出金額後列出最多與最少的支出金額。


```python
priceList=[]
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額"))
    priceSum += p
    priceList.append(p)
priceList.sort()   

min(priceList)
print(f"max:{priceList[-1]}") 
print(f"min:{priceList[0]}")
print(priceSum)    
print(priceList) 
print(max(priceList))    
print(min(priceList)) 
```

## 物件導向特性
+ 實體成員 non-static (instance) member
  + 實體屬性：每個物件各自擁有一份資料，通常在初始化流程中建立
  + 實體方法：物件提供的功能，可存取物件的屬性，需透過特定物件來操作
+ 類別成員 static (class) member
  + 類別屬性：同類別所有物件共享的屬性資料，在類別區段中宣告
  + 類別方法：不需指定物件，可由類別來操作的功能

## new
+ 繼承不可變的物件時使用
+ 可建立singleton設計模式

```python
class Singloeton:
    def __new__(cls):
        if not hasattr(cls,'instance'):
            cls.instance = super(Singloeton,cls).__new__(cls)
        return  cls.instance  
s1 = Singloeton()
s2 = Singloeton()
print(id(s1),":",id(s2))
```

# 物件實體方法


```python
class Employee:
    def __init__(self,name,salary=25000):
        self.name = name
        self.salary = salary
    def printInfo(self,title):
        print(title)
        print("name:",self.name)
        print("salary:",self.salary)
        
emp = Employee("Lucy",650000)
emp.printInfo("Manager")
```

```python
## 建立2組Student的物件 屬性有 name age score  可用init初始化
# 物件內有testScore方法 可判定 score是否及格  大於60回傳及格 小於60回傳不及格  
# 輸入的 score小於0 大於100 拋出OverflowError錯誤
# 物件內有printInfo方法 可輸出name age score 與 是否及格
# st1 = Student("Ken",18,86)
# st1.printInfo()
class Student:
    def __init__(self,name,age,score):
        self.name = name
        self.age = age
        self.score = score
    def testScore(self):
        if self.score >= 60:
            return "及格"
        else:
            return "不及格"
    def printInfo(self):
        print(self.name,self.age,self.score,self.testScore())
        
st1 = Student("ken",18,92)
print(st1.score)
st1.printInfo()
```
