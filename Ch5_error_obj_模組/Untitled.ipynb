```python
class Singloeton:
    def __new__(cls):
        if not hasattr(cls,'instance'):
            cls.instance = super(Singloeton,cls).__new__(cls)
        return  cls.instance  
s1 = Singloeton()
s2 = Singloeton()
print(id(s1),":",id(s2))
```

```python
## 2Student  name age score  init
# testScore  score  60 60  
#  score0 100 OverflowError
# printInfo name age score  
# st1 = Student("Ken",18,86)
# st1.printInfo()
class Student:
    def __init__(self,name,age,score):
        self.name = name
        self.age = age
        self.score = score
    def testScore(self):
        if self.score >= 60:
            return ""
        else:
            return ""
    def printInfo(self):
        print(self.name,self.age,self.score,self.testScore())
        
st1 = Student("ken",18,92)
print(st1.score)
st1.printInfo()
```
