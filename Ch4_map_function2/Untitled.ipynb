## 變數影響範圍
+ 函數外的變數：
  + 函數內可以顯示該變數內容
  + 不屬於函數的區域內都可以使用
+ 函數內的變數：
  + 只在函數內產生效果，不會影響函數外的變數
  + 若函數內沒有進行變數宣告而進行改變內容動作將會產生錯誤訊息!

````python
a = 5
def func_sum():
    a = 10
    print("函數內:",a)
    
print("函數外1:",a)    
func_sum()
print("函數外2:",a)
````

## list 傳送與接收
+ 傳送與接收資料可以使用 list。
+ 傳送與接收都是相同的位址上的資料。
+ list 內可以更新單一資料或著調整
   (append、extend、insert、remove、pop、del) 資料。
+ list 若以指派方式設定為新的list：
  + 資料將會儲存在新的位址。
  + 原來 list 沒有變更。

````python
def change(mylist):
    mylist = [1,2,3,4]
    print(id(mylist))
    
a1List = [10,20,30]
change(a1List)
print(id(a1List))
print(a1List)
````

````python
def change(mylist):
    mylist += [1,2,3,4]
    mylist.append(16)
    mylist[4] = 8
    print(id(mylist))
    
a1List = [10,20,30]
change(a1List)
print(id(a1List))
print(a1List)
````

# 匿名函數
+ 不需要定義函數名稱的，只需要用運算式或表達分析語法。
+ Python 使用 lambda 語法定義匿名函數。
+ 匿名函數是一個表達式/計算式，並不是一個執行流程區塊。
+ 匿名函數可以出現在一般函數不允許的地方，例如像 list 內部或函數呼叫參數的位置。

## 傳統函數 VS 匿名函數

````python
def f(x,y,z):
    return x + y + z
print(f(2,30,400))
f2 = lambda x,y,z:x+y+z
print(f2(2,30,400))
````

````python
mz = (lambda a = "T1",b = "T2",c = "T3" : a+b+c)
print(mz("v1","v2"))
````

````python
def f1(x):
    return x ** 2
def f2(x):
    return x ** 3
f3 = lambda y:y ** 4

t1 = [f1,f2,lambda y:y ** 4]
for f in t1:
    print(f(3))
````

## 匿名函數可用於條件分析表達上

# Map與Filter
+ 用法：map(function, sequence)
+ 將複合性資料逐一取出項目再傳入到 function 操作，最後以 list 作為回傳值。
+ filter( ) 函數用於過濾 list，過濾掉不符合條件的元素，返回由符合條件元素組成的新 list。
+ filter 接收兩個參數，第一個為函數，第二個為 list，list 的每個元素作為參數傳遞給函數進行分析，然後返回 True 或 False，最後將返回 True 的元素放到新list中。


# Map

````python
def m2(x):
    return x * 2
a = map(m2,[1,2,3,4])
list1 = list(a)
print(list1)
a = map(lambda x:x * 2,[1,2,3,4])
list2 = list(a)
print(list2)
````

````python
dictList = [{'name':'python','point':10},{'name':'java','point':8}]
print(type(dictList))
list1 = map(lambda x:x["name"],dictList)
print(list(list1) )
list2 = map(lambda x:x["point"] * 20,dictList)
print(list(list2) )
list3 = map(lambda x:x["name"] == "python",dictList)
print(list(list3) )

data1 = [1,2,3]
data2 = [10,20,30]
data3 = [5,4,3]
data4 = [6,7,8]
list4 = map(lambda x,y,z,a:x+y+z+a,data1,data2,data3,data4)
print(list(list4) )
````

## Filter

````python
a = [1,2,3,4,5,6]
list1 = filter(lambda x :x%2 == 0,a)
print(list(list1))
dictList = [{'name':'python','point':10},{'name':'java','point':8}]
list2 = filter(lambda x : x["name"] == "python",dictList)
print(list(list2))
````

## 亂數
+ 必須加入 import  random 語法。
+ 隨機整數：0 到 100 之間(包含100)，請加入 random.randint(0,100) 語法。
+ 隨機數值：請加入random.random ( ) 語法。
+ 隨機選取 0 到100 間的偶數(固定間隔)，請加入 random.randrange(0, 101, 2) 語法。

````python
import random
a = random.randint(0,100)#包含100
print(a)
b = random.randrange(0,100,2)#0~99 注意不包含100 的偶數
print(b)
c = random.random()#0~1  不包含1
print(c)
ranList = [20,16,10,5]
random.shuffle(ranList)
print(ranList)
a = random.sample("abcdefghij",3)
print(a)
b  = random.choice(["Apple","Java","Python","PHP"])
print(b)
````
