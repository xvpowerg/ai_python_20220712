# Map

````python
def m2(x):
    return x * 2
a = map(m2,[1,2,3,4])
list1 = list(a)
print(list1)
a = map(lambda x:x * 2,[1,2,3,4])
list2 = list(a)
print(list2)
````

````python
dictList = [{'name':'python','point':10},{'name':'java','point':8}]
print(type(dictList))
list1 = map(lambda x:x["name"],dictList)
print(list(list1) )
list2 = map(lambda x:x["point"] * 20,dictList)
print(list(list2) )
list3 = map(lambda x:x["name"] == "python",dictList)
print(list(list3) )

data1 = [1,2,3]
data2 = [10,20,30]
data3 = [5,4,3]
data4 = [6,7,8]
list4 = map(lambda x,y,z,a:x+y+z+a,data1,data2,data3,data4)
print(list(list4) )
````

## Filter

````python
a = [1,2,3,4,5,6]
list1 = filter(lambda x :x%2 == 0,a)
print(list(list1))
dictList = [{'name':'python','point':10},{'name':'java','point':8}]
list2 = filter(lambda x : x["name"] == "python",dictList)
print(list(list2))
````
